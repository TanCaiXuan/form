```python
import matplotlib.pyplot as plt
import pandas as pd

status_data = trdr_dev.split_by_approval_status()

date_col = 'cr94d_requestdate'

for df in status_data.values():
    df[date_col] = pd.to_datetime(df[date_col], errors='coerce')

for status, df in status_data.items():
    daily_counts = df.groupby(df[date_col].dt.date).size()

    plt.figure(figsize=(10, 5))
    plt.plot(daily_counts.index, daily_counts.values, marker='o')
    plt.title(f"{status.capitalize()} Records Over Time")
    plt.xlabel("Date")
    plt.ylabel("Count")
    plt.grid(True)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
import matplotlib.image as mpimg

# Global font settings
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['font.size'] = 11

# Rename mappings
request_form_rename = {
    'cr94d_requestdate': 'Request Date',
    'cr94d_affectedhosttype': 'Affected Host Type',
    'cr94d_affectedhost': 'Affected Host',
    'cr94d_affectedhostdescription': 'Affected Host Description',
    'cr94d_deviatedclause': 'Deviated Clause',
    'cr94d_deviatedpolicy': 'Deviated Policy',
    'cr94d_deviationreasonwithimpact': 'Deviation Reason with Impact',
    'cr94d_workaround': 'Workaround'
}

requestor_rename = {
    'cr94d_requestorname': 'Requestor Name',
    'cr94d_requestormail': 'Requestor Email',
    'cr94d_requestordepartmentlist': 'Requestor Department List',
    'cr94d_requestorbusinessunit': 'Requestor Business Unit',
    'cr94d_requestordesignation': 'Requestor Designation'
}

hod_rename = {
    'hoda_approvalstatus': 'HOD Approval Status',
    'cr94d_dateonapproval': 'Date on Approval',
    'cr94d_hod_name': 'HOD Name',
    'cr94d_hodmail': 'HOD Email',
    'cr94d_hodaapprovalreason': 'HOD Approval Reason'
}

# Columns to display
request_form_cols = list(request_form_rename.keys())
requestor_cols = list(requestor_rename.keys())
hod_cols = list(hod_rename.keys())

# Approval status mapping
approval_status_map = {
    850550000: 'Open',
    850550001: 'Rejected by HOD',
    850550002: 'Approved by HOD',
    850550003: 'Rejected by Tech Risk Manager',
    850550004: 'Approved by Tech Risk Manager',
    850550005: 'Rejected by ICS Head',
    850550006: 'Approved by ICS Head',
    850550007: 'Pending for the decision of GIOT Head',
    850550008: 'Approved by GIOT Head',
    850550009: 'Rejected by GIOT Head'
}

# Affected Host Type mapping
affected_host_type_map = {
    850550000: 'Information and Data',
    850550001: 'Hardware',
    850550002: 'Software',
    850550003: 'IT Infrastructure',
    850550004: 'Services',
    850550005: 'Processes',
    850550006: 'Facilities',
    850550007: 'Others',
}

# Department list mapping
department_list_map = {
    850550000: 'GIOT',
    850550001: 'Human Resource',
    850550002: 'Finance',
    850550003: 'Commercial',
    850550004: 'Safety',
    850550005: 'Production',
    850550006: 'Engineering',
    850550007: 'Others',
}

# Dynamic title positioning
def dynamic_title_y(row_count):
    return 0.85 + (0.015 * max(0, 8 - row_count))

# Render table function
def df_to_vertical_table(ax, df_row, rename_map=None):
    if rename_map is None:
        data = [[col, str(df_row[col])] for col in df_row.index]
    else:
        data = [[rename_map.get(col, col), str(df_row[col])] for col in df_row.index]

    ax.axis('off')

    tbl = ax.table(
        cellText=data,
        colLabels=["Field", "Value"],
        loc='center',
        cellLoc='left',
        bbox=[0, 0, 1, 0.75]
    )

    tbl.auto_set_font_size(False)
    tbl.set_fontsize(11)
    tbl.scale(1, 1.3)

    # Styling cells
    for key, cell in tbl.get_celld().items():
        cell.set_text_props(ha='left', va='center')
        cell.set_linewidth(0.5) #border thickness
        cell.set_edgecolor('black')

        if key[0] == 0:
            cell.set_fontsize(12)
            cell.set_text_props(weight='bold')
            cell.set_facecolor('#f0f0f0')

    ax.set_frame_on(False)

grouped = rejected_df_hoda.groupby('cr94d_techriskdeviationid_x')

with PdfPages('tech_risk_deviation_report_rejected_HOD.pdf') as pdf:
    for deviation_id, group_df in grouped:
        # Apply mapping
        group_df['cr94d_affectedhosttype'] = group_df['cr94d_affectedhosttype'].map(affected_host_type_map)
        group_df['cr94d_requestordepartmentlist'] = group_df['cr94d_requestordepartmentlist'].map(department_list_map)

        fig, axes = plt.subplots(
            nrows=3,
            figsize=(8.27, 11.69),  
            gridspec_kw={'height_ratios': [2, 1.5, 1.2]}
        )
        plt.subplots_adjust(top=0.93, hspace=0.1) 


        #img
        img = mpimg.imread('src/img/Seatrium Limited logo.png') 
        img_ax = fig.add_axes([0.75, 0.9, 0.2, 0.06], anchor='NE', zorder=1)
        #                     [left, bottom, width, height]
        img_ax.imshow(img)
        img_ax.axis('off')
        
        # Header info
        extension_rid = group_df['cr94d_extension_rid'].dropna().unique()
        extension_rid_str = extension_rid[0] if len(extension_rid) > 0 else None

        approval_status_num = group_df['cr94d_approvalstatus'].iloc[0]
        approval_status_str = approval_status_map.get(approval_status_num, f"Status {approval_status_num}")

        # Title text
        title_text = f"Tech Risk Deviation ID: {deviation_id}"
        if extension_rid_str:
            title_text += f" | Extension RID: {extension_rid_str}"
        title_text += f" | Approval Status: {approval_status_str}"

        fig.text(0.01, 0.93, title_text, fontsize=12, fontweight='bold', ha='left')

        # Underline
        fig.lines.append(
            plt.Line2D([0.01, 0.99], [0.915, 0.915], transform=fig.transFigure, color='black', linewidth=0.5)
        )

        # Take the first row
        row = group_df.iloc[0]

        # Section 1: Request Form
        axes[0].text(
            0, dynamic_title_y(len(request_form_cols)),
            'Request Form Details',
            fontsize=12, fontweight='bold', ha='left', transform=axes[0].transAxes
        )
        df_to_vertical_table(axes[0], row[request_form_cols], rename_map=request_form_rename)

        # Requestor Details
        axes[1].text(
            0, dynamic_title_y(len(requestor_cols)),
            'Requestor Details',
            fontsize=12, fontweight='bold', ha='left', transform=axes[1].transAxes
        )
        df_to_vertical_table(axes[1], row[requestor_cols], rename_map=requestor_rename)

        # HOD Approval Details
        axes[2].text(
            0, dynamic_title_y(len(hod_cols)),
            'HOD Approval Details',
            fontsize=12, fontweight='bold', ha='left', transform=axes[2].transAxes
        )
        df_to_vertical_table(axes[2], row[hod_cols], rename_map=hod_rename)

        # Save to PDF
        pdf.savefig(fig, bbox_inches='tight',dpi=300)
        plt.close(fig)

print("PDF report saved as tech_risk_deviation_report_vertical.pdf")
```
