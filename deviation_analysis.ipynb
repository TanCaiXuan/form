```python
status_tables = trdr_dev.split_by_approval_status()

print("Closed count:", len(status_tables['closed']))
print("Open count:", len(status_tables['open']))
print("Rejected count:", len(status_tables['rejected']))
```

```python
import matplotlib.pyplot as plt
import pandas as pd

status_data = trdr_dev.split_by_approval_status()

date_col = 'cr94d_requestdate'

for df in status_data.values():
    df[date_col] = pd.to_datetime(df[date_col], errors='coerce')

for status, df in status_data.items():
    daily_counts = df.groupby(df[date_col].dt.date).size()

    plt.figure(figsize=(10, 5))
    plt.plot(daily_counts.index, daily_counts.values, marker='o')
    plt.title(f"{status.capitalize()} Records Over Time")
    plt.xlabel("Date")
    plt.ylabel("Count")
    plt.grid(True)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```
